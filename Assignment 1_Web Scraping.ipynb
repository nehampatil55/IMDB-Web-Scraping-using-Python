# Web Scraping from IMDB Site

# Import the required modules.

```python
import requests
```

```python
from bs4 import BeautifulSoup
```

```python
import pandas as pd
```

Import csv module to save data in csv file

```python
import csv
```

```python
Movie_details={}
Movie_ct=0

#download the IMDB site of top 250 movies   
source = requests.get('https://www.imdb.com/chart/top/')
source.raise_for_status()

#Create soup object
soup = BeautifulSoup(source.text,'html.parser')

#Find movies using html tag 'tbody' and class'lister-list'

movies= soup.find('tbody',class_="lister-list").find_all('tr')
    
    
# Fetch the data like title,rank,released year and rating using tag 'td' and class 'titleColumn'
for movie in movies:
    
    title= movie.find('td',class_= 'titleColumn').a.text
         
       # print(title)
    
    Rank= movie.find('td',class_= 'titleColumn').get_text(strip=True).split('.')[0]
        
       #print(Rank)
        
    year = movie.find('td',class_= 'titleColumn').span.text.strip('()')
        
       #print(year)
        
    rating =  movie.find('td',class_= 'ratingColumn imdbRating').strong.text

  
    Movie_ct += 1
    Movie_details[Movie_ct]=[Rank,title,year,rating]

     
    print('Rank:',Rank, 'Movie_title',title,'Year', year,'Movie_rating', rating)
    
# Movie data is converted to pandas dataframe

#Movie_details_df= pd.DataFrame.from_dict( Movie_details, columns=['Rank','title','year','rating'])
#Movie_details_df.head(10)
    
    

```

```python
# Movie data is converted to pandas dataframe

Movie_details_df= pd.DataFrame.from_dict( Movie_details, orient='index',columns=['Rank','title','year','rating'])
#Movie_details_df.head(10)
Movie_details_df
```
